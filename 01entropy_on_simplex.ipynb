```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure(figsize=(12, 12))
ax = plt.gca(projection='3d')

eps = 1e-6
k = np.linspace(eps, 1-eps, 100)
theta = np.linspace(eps, 1-eps, 100)
k, theta = np.meshgrid(k, theta)

x = 1 - k
z = k * (1 - theta)
y = 1 - (x + z)

H = - (x*np.log(x) + y*np.log(y) + z*np.log(z))

xx = k / 2 * (1 + theta)
yy = k * 3**.5 / 2 * (1 - theta)

ax.plot_surface(xx, yy, H, cmap='viridis', alpha=.8)
ax.contourf(xx, yy, H, 50, zdir='z', offset=0)
ax.set_title('Entropy on trinomial simplex', fontsize=22)
ax.view_init(elev=22, azim=-20)
plt.show() 
```
